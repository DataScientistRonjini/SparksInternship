# **2. Outlier Handling**
#### For continuous variables

```python
plt.figure(figsize=(15,15))
parameters = ['Sales', 'Quantity', 'Discount', 'Profit']
for i in enumerate(parameters):
    plt.subplot(2,2,i[0]+1)
    sns.boxplot(df_samplestore[i[1]])
```

### There exists outliers for all the above variables so an outlier treatment is necessary.

```python
# Outlier Treatment
```

```python
# Treating outliers of column Sales
q1_samplestore = df_samplestore['Sales'].quantile(0.01)
q4_samplestore = df_samplestore['Sales'].quantile(0.99)
```

```python
df_samplestore['Sales'][df_samplestore['Sales']<=q1_samplestore] = q1_samplestore
```

```python
df_samplestore['Sales'][df_samplestore['Sales']>=q4_samplestore] = q4_samplestore
```

```python
# checking the outliers of sales data
sns.boxplot(df_samplestore['Sales'])
```

```python
# There seems continuity in the outliers so we will leave it like this.
```

```python
# Treating outliers of column Quantity
q1_samplestore_quantity = df_samplestore['Quantity'].quantile(0.01)
q4_samplestore_quantity = df_samplestore['Quantity'].quantile(0.99)
```

```python
df_samplestore['Quantity'][df_samplestore['Quantity']<=q1_samplestore_quantity] = q1_samplestore_quantity
```

```python
df_samplestore['Quantity'][df_samplestore['Quantity']>=q4_samplestore_quantity] = q4_samplestore_quantity
```

```python
# checking the outliers of quantity data
sns.boxplot(df_samplestore['Quantity'])
```

```python
# Let the outliers remain like this
```

```python
# Treating outliers of column Discount
q1_samplestore_discount = df_samplestore['Discount'].quantile(0.01)
q4_samplestore_discount = df_samplestore['Discount'].quantile(0.99)
```

```python
df_samplestore['Discount'][df_samplestore['Discount']<=q1_samplestore_discount] = q1_samplestore_discount
```

```python
df_samplestore['Discount'][df_samplestore['Discount']>=q4_samplestore_discount] = q4_samplestore_discount
```

```python
# checking the outliers of discount data
sns.boxplot(df_samplestore['Discount'])
```

```python
# Let the discount outliers remain like this.
```

```python
# Treating outliers of column Profit
q1_samplestore_profit = df_samplestore['Profit'].quantile(0.01)
q4_samplestore_profit = df_samplestore['Profit'].quantile(0.99)
```

```python
df_samplestore['Profit'][df_samplestore['Profit']<=q1_samplestore_profit] = q1_samplestore_profit
```

```python
df_samplestore['Profit'][df_samplestore['Profit']>=q4_samplestore_profit] = q4_samplestore_profit
```

```python
# checking the outliers of Profit data
sns.boxplot(df_samplestore['Profit'])
```

```python
# The outliers have got some continuity so we will keep it as it is.
```
